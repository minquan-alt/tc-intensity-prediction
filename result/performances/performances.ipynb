```python
import matplotlib.pyplot as plt
import numpy as np

def plot_mae(mae_data, horizons, metric="test"):
    plt.figure(figsize=(10,12))
    for model, data in mae_data.items():
        plt.plot(horizons, data[metric], marker='o', label=model)
    # plt.plot(horizons, df_cma_era5_tkan["MAE 2021"], label='TKAN-2021');
    # plt.plot(horizons, df_cma_era5_gru["MAE 2021"], label='GRU-2021');
    # plt.plot(horizons, df_cma_era5_lstm["MAE 2021"], label='LSTM-2021');
    plt.plot(horizons, df_cma_era5_rnn_kan_v1_1["MAE 2021"], label='RNN-KAN-V1.1');
    plt.plot(horizons, df_cma_era5_rnn_kan_v1_2["MAE 2021"], label='RNN-KAN-V1.2');

    
    plt.xlabel("Prediction Horizon (hours)")
    plt.ylabel(f"MAE ({metric})")
    plt.title(f"MAE vs Horizon - {metric}")
    plt.legend()
    plt.grid(True)
    plt.show()
    
```
